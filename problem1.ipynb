```python
import json
from urllib.request import urlopen
import pandas as pd
```

### Fetch data from http://jsonplaceholder.typicode.com/users

```python
user_data_df = pd.read_json('http://jsonplaceholder.typicode.com/users')
```

```python
user_data_df
```

### open salary_data.json file

```python
with open('JSONFILES/salary_data.json') as f:
  data = json.load(f)
  salary_data=data['array']
  salary_data_df = pd.DataFrame(salary_data)
```

```python
salary_data_df
```

### Join the fetched data with salary data from JSON file by using id

```python
user_salary_data_pd = pd.merge(user_data_df, salary_data_df)
```

```python
user_salary_data_pd
```

#### Add one field to represent salary in USD (salary in JSON file is in IDR) using currency converter (such as https://free.currencyconverterapi.com)

```python
with urlopen("https://free.currconv.com/api/v7/convert?q=USD_IDR&compact=ultra&apiKey=af6b3584ef2f21d8c446") as response:
    source = response.read()
    USDtoIDR = json.loads(source)
```

```python
user_salary_data_pd['salaryInUSD'] = user_salary_data_pd['salaryInIDR']/USDtoIDR['USD_IDR']
```

```python
user_salary_data_pd
```

#### Output from the endpoint should be: id, name, username, email, address, phone, salaryInIDR, and salaryInUSD

```python
user_salary_data__sliced_pd = user_salary_data_pd.loc[:, ['id','name','username','email','address','phone','salaryInIDR','salaryInUSD']]
```

```python
user_salary_data__sliced_pd
```

```python
user_salary_data__sliced_pd.to_json('JSONFiles/salary_data_modified.json', orient='records', indent=2)
```
