```python
import json
from urllib.request import urlopen
import pandas as pd
```

Open sensor_data.json file

```python
with open('JSONFILES/sensor_data.json') as f:
  data = json.load(f)
  sensor_data=data['array']
  sensor_data_df = pd.DataFrame(sensor_data)
```

```python
sensor_data_df
```

```python
sensor_data_df['day'] = pd.to_datetime(sensor_data_df['timestamp'], unit='ms').dt.date
```

```python
sensor_data_df
```

### Aggregate sensor data from JSON file on roomArea and by day.
### min, max, median, and avg from sensors value (temperature, humidity)
e.g. room area 1 in july 2, 2020 has {min, max, median, average} {temp, humidity} of some values

```python
result = sensor_data_df.groupby(['roomArea', 'day']).agg({'temperature': ['mean', 'min', 'max','median'],'humidity': ['mean', 'min', 'max','median'] })
```

```python
result
```

```python
result.to_json('JSONFiles/sensor_data_modified.json', orient='index', indent=2)
```
