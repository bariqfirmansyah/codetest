#### Make one program that streams sensor data (Temperature and Humidity) in at least five different rooms with 2 minutes push rate. This streaming data should be logged in one log file (in .txt, .csv, or .json format) . The program should be continuously running until the user stops the process.

```python
import sched, time, json, os
from urllib.request import urlopen
from csv import writer
import pandas as pd
```

```python
with open('JSONFiles/sensor_data.json') as f:
    sensor_data = json.load(f)["array"]
    sensor_data_df = pd.DataFrame(sensor_data)
```

```python
sensor_data_df
```

### Push and log sensor data with interval=2 minutes

```python
if os.path.isfile('JSONFiles/stream_log.csv'):
    os.remove('JSONFiles/stream_log.csv')

with open('JSONFiles/stream_log.csv', 'a', newline='') as f_object:  
    writer_object = writer(f_object)
    # log list file to csv
    header=['minute','room1Temp','room1Humid','room2Temp','room2Humid','room3Temp','room3Humid']
    # Pass the data in the list as an argument into the writerow() function
    writer_object.writerow(header)

s = sched.scheduler(time.time, time.sleep)
minute=0
counter=0
def do_something(sc): 
    global minute
    global counter
    print("Minute: "+str(minute))
    print("Room 1 Temperature: "+str(sensor_data_df.at[counter, 'temperature'])+", Room 1 Humidity: "+str(sensor_data_df.at[counter, 'humidity']))
    print("Room 2 Temperature: "+str(sensor_data_df.at[counter+1, 'temperature'])+", Room 2 Humidity: "+str(sensor_data_df.at[counter+1, 'humidity']))
    print("Room 3 Temperature: "+str(sensor_data_df.at[counter+2, 'temperature'])+", Room 3 Humidity: "+str(sensor_data_df.at[counter+2, 'humidity']))
    print('------------waiting for the next measurement------------')
    
    with open('JSONFiles/stream_log.csv', 'a', newline='') as f_object:  
        # Pass the CSV  file object to the writer() function
        writer_object = writer(f_object)
        # log list file to csv
        list_data=[minute,sensor_data_df.at[counter, 'temperature'],sensor_data_df.at[counter, 'humidity'],sensor_data_df.at[counter+1, 'temperature'],sensor_data_df.at[counter+1, 'humidity'],sensor_data_df.at[counter+2, 'temperature'],sensor_data_df.at[counter+2, 'humidity']]
        # Pass the data in the list as an argument into the writerow() function
        writer_object.writerow(list_data)
        # Close the file object
        f_object.close()
        
    minute=minute+2
    counter+=3 # because there are three rooms
    s.enter(120, 1, do_something, (sc,))
    
s.enter(0, 1, do_something, (s,))
s.run()
```
